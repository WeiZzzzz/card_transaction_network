```python
# 根据df1的数据，建立一个customerId和merchantCategoryCode的bipartite network
import networkx as nx
import matplotlib.pyplot as plt
from networkx.algorithms import bipartite

B = nx.Graph()
B.add_nodes_from(df1['customerId'], bipartite=0)
B.add_nodes_from(df1['externalParty.merchantCategoryCode'], bipartite=1)
B.add_edges_from([(row['customerId'], row['externalParty.merchantCategoryCode']) for idx, row in df1.iterrows()])
# project the bipartite network into a customer network
G = bipartite.projected_graph(B, df1['customerId'])
# plot the customer network
nx.draw(G, with_labels=False, node_size=10, font_size=8, node_color='b', edge_color='gray', width=0.5)
plt.show()
```

```python
# 统计每个客户的总消费次数
cntpercust = df1.groupby('customerId')['_id'].count().sort_values(ascending=False).reset_index()
# 筛选总消费次数大于10次的客户
cntpercust = cntpercust[cntpercust['_id'] > 20]
# 筛选出这些客户的数据
df_validcust = df1[df1['customerId'].isin(cntpercust['customerId'])]
df_validcust
```

```python
# 统计每个客户在不同的商家类别下的消费次数
df_count = df_validcust.groupby(['customerId', 'externalParty.merchantCategoryCode'])['_id'].count().unstack().fillna(0)
df_count
```

```python
df_count2 = df_validcust.groupby(['customerId', 'externalParty.merchantCategoryCode'])['_id'].count().reset_index()
df_count2.columns = ['customerId', 'merchantCategoryCode', 'count']
df_count2
```

```python
B = nx.Graph()
for index, row in df_count2.iterrows():
    B.add_node(row['customerId'], bipartite=0)
    B.add_node(row['merchantCategoryCode'], bipartite=1)
    B.add_edge(row['customerId'], row['merchantCategoryCode'], weight=row['count'])
# project the bipartite network into a customer network
# G = bipartite.projected_graph(B, set(df_count2['customerId']))
pos = nx.bipartite_layout(B, set(df_count2['customerId']))
# node size according to the degree of the node
node_size = [B.degree(node)/2 for node in B.nodes]
# plot the customer network
nx.draw(B, pos, with_labels=False, node_size=node_size, font_size=8, node_color='b', edge_color='gray', width=0.2)
```

```python
# count the number of transactions in common merchant categories for each pair of customers
df_count3 = df_validcust.groupby(['customerId', 'externalParty.merchantCategoryCode'])['_id'].count().unstack()
df_count3
```

```python
# compare each pair of customers and calculate the number of common merchant categories
import itertools
common_merchant = []
for pair in itertools.combinations(df_count3.index, 2):
    common = df_count3.loc[pair[0]].notna() & df_count3.loc[pair[1]].notna()
    common_merchant.append([pair[0], pair[1], common.sum()])
df_common = pd.DataFrame(common_merchant, columns=['customerId1', 'customerId2', 'common_merchant'])
df_common
```

```python
# construct the customer network based on the number of common merchant categories
B = nx.Graph()
for index, row in df_common.iterrows():
    B.add_edge(row['customerId1'], row['customerId2'], weight=row['common_merchant'])
# plot the customer network
pos = nx.spring_layout(B, seed=42, k=0.1, iterations=100, scale=2)
node_size = [B.degree(node)/10 for node in B.nodes]
nx.draw(B, pos, with_labels=False, node_size=node_size, font_size=8, node_color='b', edge_color='gray', width=0.2)
```
